```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

atx_crime = "crime_by_zip_csv/crime_by_zip_combined.csv"

atx_crime_df = pd.read_csv(atx_crime)
atx_crime_df.dropna(axis=0, inplace=True)
atx_crime_df = atx_crime_df.rename(columns = {"Motor-V Theft": "Auto_Theft",
                                    })
new_df = atx_crime_df.sort_values(by=["Zipcode", "Year"])
# new_df = atx_crime_df.groupby(["Zipcode", "Year"]).mean()
new_df
```

```python
crim_zip = new_df.pct_change()*100
crim_zip = crim_zip.rename(columns = {"Burglary": "Change_Burglary",
                                      "Theft": "Change_Theft",
                                      "Auto_Theft": "Change_Auto",
                                    })

crim_zip
```

```python
new_df["Change_Burglary"] = crim_zip["Change_Burglary"]
new_df["Change_Theft"] = crim_zip["Change_Theft"]
new_df["Change_Auto"] = crim_zip["Change_Auto"]
new_df
```

```python
combined_df = new_df[["Year", "Zipcode", "Burglary", "Change_Burglary", "Theft", "Change_Theft", "Auto_Theft", "Change_Auto"]]
combined_df
```
